```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This document is about makeup",
        "This document is about sports"
    ],
    metadatas=[
        {"category": "makeup"},
        {"category": "sports"}
    ],
    ids=["id1", "id2"]
)
```

```python
all_docs = collection.get()
all_docs
```

```python
documents = collection.get(ids=["id1"])
documents
```

```python
results = collection.query(
    query_texts=['This query is about fractures'],
    n_results=2
)
results
```
