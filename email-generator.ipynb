```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Access the API key from environment
groq_api_key = os.getenv("GROQ_API_KEY")

# Initialize the model
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    groq_api_key=groq_api_key
)

# Query the model
response = llm.invoke("The first woman prime minister of India")
print(response.content)
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/director-software-engineering/job/R-43982")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in 150N format containing the following keys: role experience, skills and description".
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):
    """
)
chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={"page_data": page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
  for _, row in df.iterrows():
    collection.add(documents=row["Techstack"],
                    metadatas={"links": row["Links"]},
                   ids=[str(uuid.uuid4())])
```

```python
links = collection.query(
    query_texts=["Experience in python", "Experience in React"],  n_results=2).get('metadatas', [])
links
```
